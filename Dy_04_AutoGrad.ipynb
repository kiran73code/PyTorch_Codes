```python
## PyTorch Gradient  (AutoGrad)
```

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
 y = x**3 +3*x**2
 y
```

```python
y.backward()
```

```python
x.grad
```

```python
x1 = torch.tensor([[1.,2.,3,4],[4.,3.,2.,1]] , requires_grad=True)
x1
```

```python
y1 = 3*x1 + 2
y1
```

```python
out = y1.mean()
out
```

```python
out.backward()
```

```python
x1.grad
```
